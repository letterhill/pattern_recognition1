```python
#수치형 열 - 중앙값으로 대체
numeric_cols = df.select_dtypes(include=['number']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# 문자열 열 - 최빈값으로 대체
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
df.info()
```

```python
# 문자열 숫자열로 변형

df["data_channel"] = df["data_channel"].replace({'World': 0, 'Lifestyle': 1, 'Tech': 2,
    'Entertainment': 3, 'Business': 4, 'Social Media': 5})
df["weekday"] = df["weekday"].replace({'Monday': 0, 'Tuesday': 1, 'Wednesday': 2,
    'Thursday': 3, 'Friday': 4, 'Saturday': 5, 'Sunday': 6})
```

```python
df
```

```python
df["y"] = (df["y"] > 0.5).astype(int)
```

```python
from sklearn.model_selection import train_test_split

random_state = 100
shuffle = True
test_size_ratio = 0.25

train_df, test_df = train_test_split(df, test_size=test_size_ratio, random_state=random_state, shuffle=shuffle)
print(train_df.shape, test_df.shape)

target = "y"

X_train = train_df.drop(target, axis=1).values
y_train = train_df[target].values

X_test = test_df.drop(target, axis=1).values
y_test = test_df[target].values

target_names = ["0", "1"]
feature_names = train_df.drop(target, axis=1).columns
```

```python
# 모델 훈련
from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression(solver='liblinear', random_state=random_state)
logistic_reg.fit(X_train, y_train)
```

```python
# 예측 수행
y_prob = logistic_reg.predict_proba(X_test)[:, 1]

# 임계값 기준 이진 예측값 생성
threshold = 0.5
y_pred = (y_prob >= threshold).astype(int)
```

```python
# 평가 지표 계산
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt


accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)


print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print(f"AUC: {auc}")
```
