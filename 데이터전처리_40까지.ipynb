```python
ref_cols = ['self_reference_min_shares', 'self_reference_max_shares', 'self_reference_avg_sharess']

### 1. num_self_hrefs == 1 이고 ref_cols에 결측치가 있는 경우 → 동일한 값으로 채우기
condition_1 = (df['num_self_hrefs'] == 1)

for idx in df[condition_1].index:
    values = df.loc[idx, ref_cols]
    non_na = values.dropna()
    if len(non_na) > 0:
        fill_value = non_na.iloc[0]  # 첫 번째 non-NaN 값 사용
        df.loc[idx, ref_cols] = fill_value

### 2. num_self_hrefs가 NaN이고, self_reference_* 값 조건에 따라 1로 채우기
condition_2 = df['num_self_hrefs'].isna()

for idx in df[condition_2].index:
    min_val = df.loc[idx, 'self_reference_min_shares']
    max_val = df.loc[idx, 'self_reference_max_shares']
    avg_val = df.loc[idx, 'self_reference_avg_sharess']

    if (
        pd.notna(min_val) and pd.notna(max_val) and min_val == max_val and min_val != 0
    ) or (
        pd.notna(min_val) and pd.notna(avg_val) and min_val == avg_val and min_val != 0
    ) or (
        pd.notna(max_val) and pd.notna(avg_val) and max_val == avg_val and max_val != 0
    ):
        df.loc[idx, 'num_self_hrefs'] = 1
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

def fill_null_by_nearest(df, column_pairs):
    df_filled = df.copy()

    for a_col, b_col in column_pairs:
        # a가 null이고 b는 값 있는 행 → 대체 대상
        target_rows = df_filled[df_filled[a_col].isnull() & df_filled[b_col].notnull()]

        # a와 b 모두 값이 있는 행 → 참조 대상
        reference_rows = df_filled[df_filled[a_col].notnull() & df_filled[b_col].notnull()]

        if reference_rows.empty or target_rows.empty:
            continue  # 대체할 게 없거나 참조할 게 없으면 건너뜀

        # 최근접 이웃 모델 학습 (1-NN)
        nbrs = NearestNeighbors(n_neighbors=1)
        nbrs.fit(reference_rows[[b_col]])

        # 최근접 이웃 검색
        distances, indices = nbrs.kneighbors(target_rows[[b_col]])

        # null 채우기
        for i, idx in enumerate(target_rows.index):
            matched_idx = reference_rows.iloc[indices[i][0]].name
            df_filled.at[idx, a_col] = df_filled.at[matched_idx, a_col]

    return df_filled
```

```python
# 서로 비교_대체할 것들
column_pairs = [
    ('n_unique_tokens', 'n_non_stop_unique_tokens'),
    ('n_non_stop_words', 'average_token_length'),
    ('global_sentiment_polarity', 'global_rate_positive_words'),
    ('global_sentiment_polarity', 'global_rate_negative_words'),
    ('global_rate_positive_words', 'rate_positive_words'),
    ('global_rate_negative_words', 'rate_negative_words'),

]

# 적용
df_result = fill_null_by_nearest(df, column_pairs)
```

# min, max, avg 비율 결측치 처리
* avg만 없는 row => (min+max)/2
* min만 없는 row -> avg 기준 비교 대체
* max만 없는 row -> avg 기준

a = avg-min, b = avg-max c = max-min

* min만 O -> avg-min 차이 평균 max-min 차이 평균 계산해서 더해서 넣기 (avg=min+a, max=min+a+b)
* max만 O -> 빼서 넣기 (min=max-a-b, avg=max-b)
* avg만 O -> 빼고 더해서 넣기 (min=avg-a, max=avg+b)
